```python
import sqlite3
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
db_filename = 'dinofunworld-1.db'
conn = sqlite3.connect(db_filename)
c = conn.cursor()

c.execute("SELECT a.Name,COUNT(c.visitorID) AS Num FROM checkin c,attraction a WHERE c.attraction = a.AttractionID AND a.Category LIKE '%Thrill Rides%' AND c.type LIKE '%actual%' GROUP BY a.Name")
counts = c.fetchall()

thrillRides = pd.DataFrame.from_records(counts, columns=['Name', 'Num'])

plt.pie(thrillRides['Num'],labels=thrillRides['Name'],shadow=False)
plt.axis('equal')
plt.show()
```

```python
import sqlite3
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
db_filename = 'dinofunworld-1.db'
conn = sqlite3.connect(db_filename)
c = conn.cursor()

c.execute("SELECT c.timestamp,COUNT(c.visitorID) AS Num FROM checkin c,attraction a WHERE c.attraction = a.AttractionID AND a.Name LIKE '%Atmosfear%' AND c.type LIKE '%actual%' GROUP BY c.timestamp")
counts = c.fetchall()

timeFrame = pd.DataFrame.from_records(counts,columns=['Time','Num'])

plt.plot(range(len(timeFrame['Time'])),timeFrame['Num'])
plt.show()
```
